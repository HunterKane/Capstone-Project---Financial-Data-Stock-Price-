```python
#Import datetime for stock prices datetime range
from datetime import datetime 
```

```python
#Filter start/end date
start = datetime(2006, 1, 1)
```

```python
end = datetime(2016, 1, 1)
```

# Get data for listed banks

```python
#Get info from the web 
BAC = data.DataReader("BAC", 'yahoo', start, end)
BAC
```

```python
#CitiGroup
C = data.DataReader("C", 'yahoo', start, end)
C
```

```python
#Goldman Sachs 
GS = data.DataReader("GS", 'yahoo', start, end)
GS
```

```python
#JPMorgan
JPM = data.DataReader("JPM", 'yahoo', start, end)
JPM
```

```python
#Wells Fargo
WPC = data.DataReader("WFC", 'yahoo', start, end)
WPC
```

```python
#Morgan Stanley
MS = data.DataReader("MS", 'yahoo', start, end)
MS
```

```python
#Create ticker list for each bank
tickers = ['BAC', 'JPM', 'WPC', 'C', 'GS', 'MS']
```

```python
tickers.sort()
```

```python
tickers
```

### Use pd.concat to concatenate the bank dataframes together to a single data frame called bank_stocks. Set the keys argument equal to the tickers list. 

```python
bank_stocks = pd.concat([BAC,C,GS,JPM,MS,WPC],axis = 1, keys=tickers)
```

```python
bank_stocks.head()
```

```python
#Update column names 
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```

```python
bank_stocks.head()
```

# EDA
- What is the max Close price for each bank's stock throughout the time period?

**Check out the documentation on Multi-Level Indexing and Using .xs.**

```python
#Max closing price for each stock  
bank_stocks.xs(key='Close', axis=1, level='Stock Info').max()
```

```python
# Create a new empty DataFrame called returns
returns = pd.DataFrame()
```

```python
#pct_change method
for tick in tickers:
    returns[tick+' Return'] = bank_stocks[tick]['Close'].pct_change()
```

```python
returns.head()
```

```python
# Create pairplot with Seabron / returns dataframe
sns.pairplot(returns[1:])
```

__Conclusion From Data__:
    
- CitiGroup was effected the most due to the world financial crisis that happened. With a very sharp spike up indicates a loss for a prolound time.

Recreate the described plots below, seaborn, matplotlib, plotly and cufflinks, or just pandas.

## Moving Averages

Let's analyze the moving averages for these stocks in the year 2008:

__Plot the rolling 30 day average against the Close Price for BAC in 2008.__

```python
#Plot 30 day moving average
plt.figure(figsize = (12,4))
BAC['Close'].loc['2008-01-01': '2009-01-01'].rolling(window=30).mean().plot(label='30 Day Moving Average')
BAC['Close'].loc['2008-01-01': '2009-01-01'].plot(label='BAC Close')
plt.legend()
```

```python
#Create a heatmap to show correlation 
sns.heatmap(bank_stocks.xs(key='Close',axis=1, level='Stock Info').corr(),annot = True)
```

```python
#Cluster map
sns.clustermap(bank_stocks.xs(key='Close',axis=1, level='Stock Info').corr(),annot = True)
```
